```python
import cv2
import numpy as np

# Cargar imagen
img = cv2.imread("capture.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = cv2.GaussianBlur(gray, (3,3), 0)

# Detección de bordes
edges = cv2.Canny(gray, 50, 150, apertureSize=3)

# Detección de líneas con Hough probabilístico
lines = cv2.HoughLinesP(
    edges,
    rho=1,
    theta=np.pi/180,
    threshold=100,
    minLineLength=100,
    maxLineGap=10
)

# Crear una copia para dibujar las líneas detectadas
hough_img = img.copy()

# Dibujar las líneas detectadas
if lines is not None:
    for line in lines:
        x1, y1, x2, y2 = line[0]
        cv2.line(hough_img, (x1, y1), (x2, y2), (0, 0, 255), 2)  # Rojo

# Mostrar resultados
cv2.imshow("Líneas detectadas por Hough", hough_img)
cv2.imshow("Bordes", edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
